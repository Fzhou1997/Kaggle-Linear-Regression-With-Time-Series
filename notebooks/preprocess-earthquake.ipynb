```python
import pandas as pd
```

```python
pd.set_option('future.no_silent_downcasting', True)
```

```python
dates_df = pd.read_csv('../data/store-sales-time-series/train.csv', usecols=['date'], parse_dates=['date'])
dates_df = dates_df.drop_duplicates()
dates_df = dates_df.sort_values(by=['date'])
dates_df = dates_df.reset_index(drop=True)
dates_df
```

```python
min_date = dates_df['date'].min()
max_date = dates_df['date'].max()
date_range = pd.date_range(start=min_date, end=max_date)
dates_df = pd.DataFrame(date_range, columns=['date'])
dates_df
```

```python
earthquake_df = pd.DataFrame()
earthquake_df['is_earthquake'] = dates_df['date'] == pd.to_datetime('2016-04-16')
earthquake_df
```

```python
for i in range(1, 31):
    earthquake_df[f'is_earthquake_{i}'] = earthquake_df['is_earthquake'].shift(i)
earthquake_df.fillna(False, inplace=True)
earthquake_df
```

```python
earthquake_df.to_csv('../data/processed/earthquake.csv', index=False)
```

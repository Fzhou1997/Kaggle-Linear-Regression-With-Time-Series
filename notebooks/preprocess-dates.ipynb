```python
import pandas as pd
import numpy as np
```

```python
dates_df = pd.read_csv('../data/store-sales-time-series/train.csv', usecols=['date'], parse_dates=['date'])
dates_df = dates_df.drop_duplicates()
dates_df = dates_df.sort_values(by=['date'])
dates_df = dates_df.reset_index(drop=True)
dates_df
```

```python
min_date = dates_df['date'].min()
max_date = dates_df['date'].max()
date_range = pd.date_range(start=min_date, end=max_date)
dates_df = pd.DataFrame(date_range, columns=['date'])
dates_df
```

```python
dates_df['day'] = dates_df.index
dates_df
```

```python
day_of_week_df = pd.DataFrame()
day_of_week_df['day_of_week'] = dates_df['date'].dt.dayofweek
day_of_week_df
```

```python
day_of_week_one_hot_df = pd.get_dummies(day_of_week_df['day_of_week'])
day_of_week_one_hot_df.columns = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_of_week_one_hot_df
```

```python
day_of_week_cyclic_df = pd.DataFrame()
day_of_week_cyclic_df['day_of_week_sin'] = np.sin(2 * np.pi * day_of_week_df['day_of_week'] / 7)
day_of_week_cyclic_df['day_of_week_cos'] = np.cos(2 * np.pi * day_of_week_df['day_of_week'] / 7)
day_of_week_cyclic_df
```

```python
day_of_month_df = pd.DataFrame()
day_of_month_df['day_of_month'] = dates_df['date'].dt.day - 1
day_of_month_df
```

```python
day_of_month_one_hot_df = pd.get_dummies(day_of_month_df['day_of_month'])
day_of_month_one_hot_df.columns = [f'Day {i}' for i in range(1, 32)]
day_of_month_one_hot_df
```

```python
day_of_month_cyclic_df = pd.DataFrame()
day_of_month_cyclic_df['day_of_month_sin'] = np.sin(2 * np.pi * day_of_month_df['day_of_month'] / 31)
day_of_month_cyclic_df['day_of_month_cos'] = np.cos(2 * np.pi * day_of_month_df['day_of_month'] / 31)
day_of_month_cyclic_df
```

```python
month_of_year_df = pd.DataFrame()
month_of_year_df['month_of_year'] = dates_df['date'].dt.month - 1
month_of_year_df
```

```python
month_of_year_one_hot_df = pd.get_dummies(month_of_year_df['month_of_year'])
month_of_year_one_hot_df.columns = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
month_of_year_one_hot_df
```

```python
month_of_year_cyclic_df = pd.DataFrame()
month_of_year_cyclic_df['month_of_year_sin'] = np.sin(2 * np.pi * month_of_year_df['month_of_year'] / 12)
month_of_year_cyclic_df['month_of_year_cos'] = np.cos(2 * np.pi * month_of_year_df['month_of_year'] / 12)
month_of_year_cyclic_df
```

```python
quarter_of_year_df = pd.DataFrame()
quarter_of_year_df['quarter_of_year'] = dates_df['date'].dt.quarter - 1
quarter_of_year_df
```

```python
quarter_of_year_one_hot_df = pd.get_dummies(quarter_of_year_df['quarter_of_year'])
quarter_of_year_one_hot_df.columns = [f'Q{i + 1}' for i in range(len(quarter_of_year_one_hot_df.columns))]
quarter_of_year_one_hot_df
```

```python
quarter_of_year_cyclic_df = pd.DataFrame()
quarter_of_year_cyclic_df['quarter_of_year_sin'] = np.sin(2 * np.pi * quarter_of_year_df['quarter_of_year'] / 4)
quarter_of_year_cyclic_df['quarter_of_year_cos'] = np.cos(2 * np.pi * quarter_of_year_df['quarter_of_year'] / 4)
quarter_of_year_cyclic_df
```

```python
year_df = pd.DataFrame()
year_df['year'] = dates_df['date'].dt.year - min_date.year
year_df
```

```python
year_one_hot_df = pd.get_dummies(year_df['year'])
year_one_hot_df.columns = [min_date.year + i for i in range(len(year_one_hot_df.columns))]
year_one_hot_df
```

```python
dates_df = pd.concat([
    dates_df, 
    day_of_week_df, 
    day_of_week_one_hot_df, 
    day_of_week_cyclic_df, 
    day_of_month_df,
    day_of_month_one_hot_df, 
    day_of_month_cyclic_df, 
    month_of_year_df,
    month_of_year_one_hot_df, 
    month_of_year_cyclic_df, 
    quarter_of_year_df,
    quarter_of_year_one_hot_df, 
    quarter_of_year_cyclic_df, 
    year_df,
    year_one_hot_df], axis=1)
dates_df
```

```python
dates_df.to_csv('../data/processed/dates.csv', index=False)
```

```python
import pandas as pd
```

```python
pd.set_option('future.no_silent_downcasting', True)
```

```python
dates_df = pd.read_csv('../data/store-sales-time-series/train.csv', usecols=['date'], parse_dates=['date'])
dates_df = dates_df.drop_duplicates()
dates_df = dates_df.sort_values(by=['date'])
dates_df = dates_df.reset_index(drop=True)
dates_df
```

```python
min_date = dates_df['date'].min()
max_date = dates_df['date'].max()
date_range = pd.date_range(min_date, max_date)
dates_df = pd.DataFrame(date_range, columns=['date'])
dates_df
```

```python
payday_df = pd.DataFrame()
payday_df['is_payday'] = (dates_df['date'].dt.day == 15) | (dates_df['date'].dt.is_month_end)
payday_df
```

```python
payday_df['is_payday_lag_1'] = payday_df['is_payday'].shift(1)
payday_df['is_payday_lag_2'] = payday_df['is_payday'].shift(2)
payday_df['is_payday_lag_3'] = payday_df['is_payday'].shift(3)
payday_df['is_payday_lag_4'] = payday_df['is_payday'].shift(4)
payday_df['is_payday_lag_5'] = payday_df['is_payday'].shift(5)
payday_df['is_payday_lag_6'] = payday_df['is_payday'].shift(6)
payday_df.fillna(False, inplace=True)
payday_df
```

```python
payday_df.to_csv('../data/processed/paydays.csv', index=False)
```
